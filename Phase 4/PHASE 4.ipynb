<B> MERGING THE ABOVE DATASETS!!<BR>

```python
cluster=pd.merge(district_wise,districtwise_sctot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year']).merge(districtwise_sttot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year']).merge(districtwise_childtot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year']).merge(districtwise_womentot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year'])
```

<B> GROUPING BY STATE/UT,DISTRICT AND SUMMING ALL THE CRIMES OVER THE YEARS!!

```python
cluster=cluster.groupby(['STATE/UT','DISTRICT']).sum().reset_index().drop(['Year'],axis=1)
```

```python
cluster
```

CLUSTERING BY K MEANS

```python
x=cluster.iloc[:,2:]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3,random_state=0)
```

```python
y_kmeans=kmeans.fit_predict(x)
```

Crime Analysis in India (2011)
India, a country with over 1.38 billion people, faces significant crime-related challenges. Data from the National Crime Records Bureau (NCRB) for 2011 provides a comprehensive overview of the crime situation in the country. In 2011, there were 2.27 million reported crime cases, representing a 4.54% decrease from the previous year.

Crimes Against Women
Crimes against women, including rape, sexual harassment, and domestic violence, are particularly alarming. In 2011, there were 213,949 reported cases of crimes against women. Of these, 11.04% were related to rape or sexual assault. Despite stricter laws and increased public awareness, these crimes remain pervasive.

Crimes Against Children
Crimes against children, such as rape, kidnapping, and murder, continue to be a critical concern. In 2011, there were 33,049 reported cases of crimes against children. Notably, 21.51% of these cases involved rape or sexual assault, and 46.24% were related to murder. From 2001 to 2011, crimes against children increased by 254.86%. Madhya Pradesh recorded the highest number of cases (40,217), followed by Uttar Pradesh (29,969) and Maharashtra (27,845).

Juvenile Crime
Juvenile involvement in crime is another serious issue. Juveniles, often misguided and vulnerable, are easily influenced towards criminal activities. In 2011, there were 27,471 recorded cases involving juveniles, up from 23,597 in 2001. Madhya Pradesh had the highest number of juvenile crime cases (61,939), while Manipur had the least.

Human Rights Violations by Police
Human rights violations by the police are a significant concern. In 2009, the highest number of cases in any single year were recorded (664 cases). From 2001 to 2011, there was an 82.815% decrease in these cases. Chhattisgarh had the highest number of human rights violations by police during this period.

Assaults on Police Personnel
Crimes against police personnel, including assaults by mobs, extremists, and during operations, are notable. Uttar Pradesh recorded the highest number of such cases (2,150) from 2001 to 2011. Jammu & Kashmir, which had 300 cases in 2001, saw an 83.33% decrease to 50 cases in 2010.

Property Crimes (Dacoity, Robbery, Burglary, Theft)
Maharashtra, Uttar Pradesh, and Madhya Pradesh reported the highest numbers of combined dacoity, robbery, burglary, and theft cases. Maharashtra had 717,452 cases, Madhya Pradesh had 389,927, and Uttar Pradesh had 366,166.

Conclusion
The data indicates an increase in reported crimes, suggesting ongoing challenges. The government must continue improving law enforcement, raising public awareness, and addressing socio-economic factors contributing to crime.
